```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#no need to write plot.show() explicitly
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
z=np.random.rand(12)
print(x)
print(y)
print(f"z={z}")
```

```python
##scatter:scatters values of x and y
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph')
plt.savefig('mat.png')
```

```python
#plot function:gives a st line
y=x*x
# plt.plot(x,y,'r--')
# plt.plot(x,y,'r*-')
# plt.plot(x,y,'ro-')
plt.plot(x,y,'r*',linestyle='dashed',linewidth=1,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
# creating subplots:plots under same canvas
plt.subplot(2,2,1)
x=y
plt.plot(x,y,'r*-')

plt.subplot(2,2,2)
y=x*x
plt.plot(x,y,'g--')

plt.subplot(2,2,3)
y=x/2
plt.plot(x,y,'bo-')

plt.subplot(2,2,4)
y=x**3
plt.plot(x,y,'r*-')
```

```python
# trignomentric functions
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave form')
plt.xlabel('x')
plt.ylabel('sin x')
plt.plot(x,y,'b')
plt.show()
```

```python
# bar graph
x=[1,2,3]
y=[10,20,30]
x2=[7,8,9]
y2=[80,90,100]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
# histogram gives density or count on the y axis
# gives density/count of numbers falling in certain range
# eg arr=(11,22,333,12,13,19,23,44,54,66,71,54)
# elements falling in range 0-20=11,13,12,19=       4(density)
```

```python
a=np.array([11,22,333,12,13,19,23,44,54,66,71,54])
plt.hist(a,bins=30)
```

```python
label='py','c','ruby','java'
sizes=[215,130,245,210]
colors=['red','green','yellow','aqua']
explode=(.1,0,0,0)
plt.pie(sizes,explode=explode,labels=label,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
# autopct:all percentages will be floating form
```
