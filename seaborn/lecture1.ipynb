```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tips[tips["sex"]=="Male"]
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="sex",style="time",size="size")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind='scatter',hue="sex",style="time",size="size")
```

```python
gap=px.data.gapminder()
temp_df=gap[gap["country"]=="India"]
temp_df
```

```python
sns.lineplot(data=temp_df,x="year",y="lifeExp")
```

```python
sns.relplot(data=temp_df,x="year",y="lifeExp",kind="line",hue="continent")
```

```python
temp=gap[gap["country"].isin(["India","Brazil","Germany"])]
sns.relplot(data=temp,x="year",y="lifeExp",kind="line",hue="country",style="continent")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",kind="line",col="sex",row="day")
```

```python
sns.relplot(data=gap,x="lifeExp",y="gdpPercap",kind="scatter",col="year",col_wrap=5)
```
