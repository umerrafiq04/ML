```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv("Algerian_forest_fires_dataset.csv",header=0)
dataset.head()
```

```python
dataset.info()
```

data cleaning


```python
# missing values
dataset.isnull().sum()
dataset[dataset.isnull().any(axis=1)]
```

```python
# we have two regions so we will label them as 0 and 1 respectively
dataset.loc[:123,"region"]=0
dataset.loc[123:,"region"]=1
```

```python
dataset
```

```python
df=dataset
df.info()
df[["region"]]=df[["region"]].astype(int)
df.info()
```

```python
# removing null values
df=df.dropna().reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# remove newly 122 indexed row
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
# look at classes
```

```python
df.info()
# this should be numerical of float and int values... not object lets typecast...
```

```python
# change the required columns as integer datatype
df.head()
```

```python
# till ws all are integre
# df.iloc[:,:6]=df.iloc[:,:6].astype(int)   #cant do like this
df[["month","year","day","Temperature","RH","Ws"]]=df[["month","year","day","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

```python
# changing other columns to float datatype
objects=[features for features in df.columns if df[features].dtype=="O"]
objects
```

```python

for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#save the cleaned dataset
df.to_csv("Algerian_forest_fires_cleaned_dataset.csv",index=False)
```
