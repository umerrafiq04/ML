```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from  sklearn.datasets import make_classification
# create dataset
X, y = make_classification(n_samples=1000, n_features=10,n_classes=2,random_state=1)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)

#from sklearn.linear_model import LogisticRegression
# logistic=LogisticRegression()
# logistic.fit(X_train,y_train) 
```

```python
# dummy model with default 0 as output
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

you can do hyperparamter tunning here to get better results


```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train) 
```

```python
# pred based on probabality
model_prob=model.predict_proba(X_test)
```

```python
# pos outcome
# model_prob=model_prob[:,1]
model_prob
# [7.56074156e-01, 2.43925844e-01]
#  prob for -ve ,+ve
```

```python
model_prob=model_prob[:,1]
model_prob
```

```python
from sklearn.metrics import roc_auc_score

# Assuming y_test, dummy_model_prob, and model_prob are defined
dummy_model_auc = roc_auc_score(y_test, dummy_model_prob)
model_auc = roc_auc_score(y_test, model_prob)

print("Dummy Model AUC: ", dummy_model_auc)
print("Model AUC: ", model_auc)
```
