```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Algerian_forest_fires_dataset.csv")
df.head()
# TEMP is the output feature
```

```python
df.info()
```

data cleaning

```python
df.isnull().sum()
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
# df.loc[123]
df[['Region']]=df[["Region"]].astype(int)
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# typecasting object datatype to int
df[["month","year","day","Temperature","RH","Ws"]]=df[["month","year","day","Temperature","RH","Ws"]].astype(int)
df.info()
```

```python
# typecasting other object datatype to float execept classes
objects=[features for features in df.columns if df[features].dtype=="O"]
objects
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
df.info()
```

```python
df.describe()
```

```python
df.to_csv("Algerian_forest_fires_cleaned_dataset.csv",index=False)
```

```python
# monthly fire analysis for region 0
dftemp=df.loc[df["Region"]==0]
dftemp
```
